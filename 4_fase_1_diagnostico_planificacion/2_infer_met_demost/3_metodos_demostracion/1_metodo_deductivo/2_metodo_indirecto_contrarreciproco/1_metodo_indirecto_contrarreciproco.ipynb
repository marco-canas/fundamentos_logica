## 1. TÃ­tulo de la Clase: **"El poder del reverso: Demostrando implicaciones mediante el contrarrecÃ­proco"**



### IntroducciÃ³n  

En este mÃ©todo buscamos probar una implicaciÃ³n $H \to T$ pero la probamos de manera indirecta, es decir, la que probamos a a su implicaciÃ³n equivalente 

$$ \neg T \to \neg H $$

Recordemos todo lo que hemos visto con el contrarrecÃ­proco. 

1. Las cuatro implicaciones asiciadas: $p \to q$, $q \to p$, $\neg p \to \neg q$, y $\neg q \to \neg p$. 
2. Modus Tollendo Tollens $[(p \to q) \land \neg q] \to \neg p$ 

## Recordar 

* ImplicaciÃ³n: $p \to q$ significa que si $p$ es verdadero, entonces $q$ tambiÃ©n es verdadero.
* $q \to p$: esta es el recÃ­proco de la implicaciÃ³n anterior 
* $\neg p \to \neg q$: la implicaciÃ³n "contraria" de la implicaciÃ³n original $p \to q$. 
* $\neg q \to \neg p $: esta es la contrarrecÃ­proca o contrapositiva de la implicaciÃ³n original $p \to q$.

### Argumentos por contraposiciÃ³n   

 Una segunda forma de argumentaciÃ³n indirecta, *argumento por contraposiciÃ³n*, se basa en 
la equivalencia lÃ³gica entre un enunciado y su contrapositivo. 

Para probar un enunciado por contraposiciÃ³n, se toma el contrapositivo del enunciado, se demuestra el contrapositivo en una demostraciÃ³n directa y se concluye que el enunciado original es verdadero. 

El razonamiento  subyacente es que, dado que un enunciado condicional es lÃ³gicamente equivalente a su contra
positivo, si el contrapositivo es verdadero, entonces el enunciado tambiÃ©n debe ser verdadero.

# Uso de la notaciÃ³n de cuantificadores 

$$ (\text{Contextos})(\text{FunciÃ³n proposicional}) $$

$$ (\forall n \in \mathbb{N})(n^{2} \quad \text{impar} \implies n \quad \text{impar} ) $$

# MÃ©todo de demostraciÃ³n por contraposiciÃ³n  

 1. Exprese el enunciado a demostrar en la forma
 $$ \forall x \ \text{en} \ D, \ \text{si} \ P(x), \ \text{entonces} \ Q(x). $$
   (Este paso se puede hacer mentalmente.)
 2. Reescriba este enunciado en forma contrapositiva
 $$ \forall x \ \text{en} \ D, \ \text{si} \ Q(x) \ \text{es falso, entonces} \ P(x) \ \text{es falso.} $$
   (Este paso tambiÃ©n se puede hacer mentalmente.)
 3. Demuestre el enunciado contrapositivo con una demostraciÃ³n directa,  
    a. Supongamos que $x$ es un elemento (particular, pero elegido arbitrariamente) de $D$ tal que $Q(x)$ es falso.  
    b. Demuestre que $P(x)$ es falso

**SoluciÃ³n** 

Primero forma que el contrapositivo del enunciado a ser probado.
 Contrapositivo: Para todo entero n, si n no es par entonces n2 no es par.
 Por el teorema del cociente-residuo con d à¸ 2, cualquier nmero entero es par o impar,

ningn entero puede ser a la vez par e impar. AsÃ­ que si un nmero entero es impar, entonces

 Contrapositivo: Para todos los enteros n, si n es impar, entonces n2 es impar.
Un cÃ¡lculo sencillo es el corazÃ³n de una demostraciÃ³n directa de este enunciado, como se

## ProposiciÃ³n 4.6.4
 Para todos los enteros $n$, si $n^{2}$ es par entonces $n$ es par.  

 **DemostraciÃ³n (por contraposiciÃ³n):**  
 
 Supongamos que n es un entero impar. [Debemos demostrar que n2 es impar.] Por 
definici n de impar, n à¸ 2k à¸ 1 para algn entero k. Por sustituci n y lgebra,
 n2 à¸ (2k à¸ 1)2 à¸ 4k 2 à¸ 4k à¸ 1 à¸ 2(2k 2 à¸ 2k) à¸ 1.
 Pero 2k2 à¸ 2k es un nmero entero, porque los productos y las sumas de los nmeros 
enteros son nmeros enteros. AsÃ­ n2 à¸ 2 à¸ (un entero) à¸ 1 y por tanto, por definici n, 
de impar, n2 es impar [que era lo que se querÃ­a demostrar].

### 7.3 En Python/Jupyter Notebook


````python


import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import ipywidgets as widgets
from sympy import symbols, Implies, Not
from sympy.logic.boolalg import Or, And, Not, Xor, Nand, Nor, truth_table
from sympy.logic.boolalg import simplify_logic
# Widget interactivo
P, Q = symbols('P Q')
def mostrar_equivalencia(P_val, Q_val):
    directo = Implies(P, Q).subs({P: P_val, Q: Q_val})
    contrar = Implies(Not(Q), Not(P)).subs({P: P_val, Q: Q_val})
    print(f"Directo: {directo} | ContrarrecÃ­proco: {contrar}")

widgets.interact(mostrar_equivalencia,
                 P_val=widgets.Dropdown(options=[True, False], description='P:'),
                 Q_val=widgets.Dropdown(options=[True, False], description='Q:'));

# AnimaciÃ³n
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

P_point, = ax.plot([], [], 'bo', markersize=15)
Q_point, = ax.plot([], [], 'go', markersize=15)
arrow1 = ax.annotate('', xy=(0,0), xytext=(0,0), arrowprops=dict(arrowstyle='->'))
arrow2 = ax.annotate('', xy=(0,0), xytext=(0,0), arrowprops=dict(arrowstyle='->', color='red'))

def animate(i):
    P_point.set_data([2], [5 + 0.1*i if i < 30 else 8])
    Q_point.set_data([8], [5])
    
    if i > 40:
        arrow1.set_position((2.5, 8))
        arrow1.xy = (7.5, 5)
        
    if i > 60:
        arrow2.set_position((7.5, 4))
        arrow2.xy = (2.5, 1)
    
    return P_point, Q_point, arrow1, arrow2

HTML(FuncAnimation(fig, animate, frames=100, interval=100).to_jshtml())

````
