## 1. Título de la Clase: **"El poder del reverso: Demostrando implicaciones mediante el contrarrecíproco"**



### Introducción  

En este método buscamos probar una implicación $H \to T$ pero la probamos de manera indirecta, es decir, la que probamos a a su implicación equivalente 

$$ \neg T \to \neg H $$

Recordemos todo lo que hemos visto con el contrarrecíproco. 

1. Las cuatro implicaciones asiciadas: $p \to q$, $q \to p$, $\neg p \to \neg q$, y $\neg q \to \neg p$. 
2. Modus Tollendo Tollens $[(p \to q) \land \neg q] \to \neg p$ 

## Recordar 

* Implicación: $p \to q$ significa que si $p$ es verdadero, entonces $q$ también es verdadero.
* $q \to p$: esta es el recíproco de la implicación anterior 
* $\neg p \to \neg q$: la implicación "contraria" de la implicación original $p \to q$. 
* $\neg q \to \neg p $: esta es la contrarrecíproca o contrapositiva de la implicación original $p \to q$.

### Argumentos por contraposición   

 Una segunda forma de argumentación indirecta, *argumento por contraposición*, se basa en 
la equivalencia lógica entre un enunciado y su contrapositivo. 

Para probar un enunciado por contraposición, se toma el contrapositivo del enunciado, se demuestra el contrapositivo en una demostración directa y se concluye que el enunciado original es verdadero. 

El razonamiento  subyacente es que, dado que un enunciado condicional es lógicamente equivalente a su contra
positivo, si el contrapositivo es verdadero, entonces el enunciado también debe ser verdadero.

# Uso de la notación de cuantificadores 

$$ (\text{Contextos})(\text{Función proposicional}) $$

$$ (\forall n \in \mathbb{N})(n^{2} \quad \text{impar} \implies n \quad \text{impar} ) $$

# Método de demostración por contraposición  

 1. Exprese el enunciado a demostrar en la forma
 $$ \forall x \ \text{en} \ D, \ \text{si} \ P(x), \ \text{entonces} \ Q(x). $$
   (Este paso se puede hacer mentalmente.)
 2. Reescriba este enunciado en forma contrapositiva
 $$ \forall x \ \text{en} \ D, \ \text{si} \ Q(x) \ \text{es falso, entonces} \ P(x) \ \text{es falso.} $$
   (Este paso también se puede hacer mentalmente.)
 3. Demuestre el enunciado contrapositivo con una demostración directa,  
    a. Supongamos que $x$ es un elemento (particular, pero elegido arbitrariamente) de $D$ tal que $Q(x)$ es falso.  
    b. Demuestre que $P(x)$ es falso

### Ejemplo 4.6.4 Si el cuadrado de un nmero entero es par, entonces el entero es par  

**Demuestre que para todo entero n, si n2 es par entonces n es par.**   

**Solución** 

Primero forma que el contrapositivo del enunciado a ser probado.
 Contrapositivo: Para todo entero n, si n no es par entonces n2 no es par.
 Por el teorema del cociente-residuo con d ฀ 2, cualquier nmero entero es par o impar, 
por lo que cualquier nmero entero que no es par es impar. También por el teorema 4.6.2, 
ningn entero puede ser a la vez par e impar. Así que si un nmero entero es impar, entonces 
no es par. Así, el contrapositivo se puede replantear de la siguiente manera:
 Contrapositivo: Para todos los enteros n, si n es impar, entonces n2 es impar.
Un cálculo sencillo es el corazón de una demostración directa de este enunciado, como se 
muestra a continuación  

## Proposición 4.6.4
 Para todos los enteros $n$, si $n^{2}$ es par entonces $n$ es par.  

 **Demostración (por contraposición):**  
 
 Supongamos que n es un entero impar. [Debemos demostrar que n2 es impar.] Por 
definici n de impar, n ฀ 2k ฀ 1 para algn entero k. Por sustituci n y lgebra,
 n2 ฀ (2k ฀ 1)2 ฀ 4k 2 ฀ 4k ฀ 1 ฀ 2(2k 2 ฀ 2k) ฀ 1.
 Pero 2k2 ฀ 2k es un nmero entero, porque los productos y las sumas de los nmeros 
enteros son nmeros enteros. Así n2 ฀ 2 ฀ (un entero) ฀ 1 y por tanto, por definici n, 
de impar, n2 es impar [que era lo que se quería demostrar].

````python


import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import ipywidgets as widgets
from sympy import symbols, Implies, Not
from sympy.logic.boolalg import Or, And, Not, Xor, Nand, Nor, truth_table
from sympy.logic.boolalg import simplify_logic
# Widget interactivo
P, Q = symbols('P Q')
def mostrar_equivalencia(P_val, Q_val):
    directo = Implies(P, Q).subs({P: P_val, Q: Q_val})
    contrar = Implies(Not(Q), Not(P)).subs({P: P_val, Q: Q_val})
    print(f"Directo: {directo} | Contrarrecíproco: {contrar}")

widgets.interact(mostrar_equivalencia,
                 P_val=widgets.Dropdown(options=[True, False], description='P:'),
                 Q_val=widgets.Dropdown(options=[True, False], description='Q:'));

# Animación
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

P_point, = ax.plot([], [], 'bo', markersize=15)
Q_point, = ax.plot([], [], 'go', markersize=15)
arrow1 = ax.annotate('', xy=(0,0), xytext=(0,0), arrowprops=dict(arrowstyle='->'))
arrow2 = ax.annotate('', xy=(0,0), xytext=(0,0), arrowprops=dict(arrowstyle='->', color='red'))

def animate(i):
    P_point.set_data([2], [5 + 0.1*i if i < 30 else 8])
    Q_point.set_data([8], [5])
    
    if i > 40:
        arrow1.set_position((2.5, 8))
        arrow1.xy = (7.5, 5)
        
    if i > 60:
        arrow2.set_position((7.5, 4))
        arrow2.xy = (2.5, 1)
    
    return P_point, Q_point, arrow1, arrow2

HTML(FuncAnimation(fig, animate, frames=100, interval=100).to_jshtml())

````

### **Actividad de Clase: Expresar Ideas, Dudas y Emociones de Manera Respetuosa y Asertiva**



**Objetivo:**  
Desarrollar habilidades de comunicación efectiva en los estudiantes para que puedan expresar sus ideas, dudas y emociones de forma respetuosa y asertiva.



**Duración:**  
15-20 minutos

---



### **1. Introducción (5 minutos)**  
- **Tema:** Introducir a los estudiantes al concepto de la comunicación asertiva, explicando cómo expresar pensamientos y emociones de manera clara y respetuosa sin agredir o ser pasivo.
- **Dinámica:** Mostrar dos ejemplos contrastantes: uno de comunicación pasiva/agresiva y otro de comunicación asertiva. Preguntar a los estudiantes cuál creen que es más efectivo y por qué.



**Materiales:**  
- Pizarra o diapositivas con ejemplos de frases pasivas, agresivas y asertivas.
  
**Ejemplo:**  
- **Pasivo:** “Bueno… no sé si es una buena idea, pero… tal vez podrías intentarlo.”
- **Agresivo:** “¡Tu idea es terrible, no va a funcionar!”
- **Asertivo:** “Creo que podríamos mejorar esa idea si consideramos otro enfoque.”

---



### **2. Ejercicio en Parejas (10 minutos)**  
- **Instrucciones:** Dividir a los estudiantes en parejas. Darles diferentes situaciones cotidianas en las que deben comunicarse (pedir ayuda, expresar desacuerdo, compartir una idea, etc.). Cada pareja debe crear una breve conversación donde practiquen la comunicación asertiva.
  
**Situaciones propuestas:**
  - Expresar desacuerdo con un compañero sobre una tarea de grupo.
  - Pedirle ayuda al profesor sobre un concepto difícil.
  - Expresar frustración por un malentendido en clase.

- **Reglas:** Cada estudiante debe asegurarse de usar lenguaje respetuoso y asertivo para expresar su posición y emociones. Al final, cada pareja debe compartir con la clase un ejemplo de lo que discutieron.

---



### **3. Retroalimentación y Discusión (5 minutos)**  
- **Discusión:** Invitar a algunos estudiantes a compartir sus experiencias de la actividad. 
- **Preguntas:** ¿Cómo se sintieron al intentar ser asertivos? ¿Les resultó difícil? ¿Cómo fue recibido el mensaje?
  
**Consejo:** Resaltar la importancia de practicar la asertividad como una herramienta para mejorar la comunicación interpersonal y reducir conflictos.

---



### **Cierre**  
- Resumir las características de la comunicación asertiva: uso de "yo" en lugar de "tú" para evitar culpabilizar, tono de voz calmado, y respeto hacia el interlocutor.
- **Tarea opcional:** Pedirles que practiquen la comunicación asertiva en sus conversaciones durante la semana y escriban una reflexión breve sobre cómo les fue.



**Material adicional:**  
- Video corto o recurso digital sobre la importancia de la comunicación asertiva.
