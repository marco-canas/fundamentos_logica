
# **4. Contenido de la Clase**  


 «Lógica» es una palabra de origen griego: Logos (λoγoς), que significa razón, reflexión, pensamiento. 

Actualmente se utiliza la palabra «lógica» para designar el estudio  formal de los razonamientos y las conclusiones a las que conducen tales razonamientos. 



En matemáticas ha resultado extremadamente útil tener una manera formal de establecer conclusiones a partir de hechos y combinaciones de hechos, porque se evitan así  muchas ambigüedades y malos entendidos.


Se dice que la lógica es un «sistema formal» porque está constituido por:   

* un lenguaje formal, 
* reglas gramaticales y 
* reglas de inferencia.   
 
Cada uno de los constituyentes del sistema se debe establecer de antemano, para construir razonamientos a partir de
 los hechos más básicos y fundamentales, hasta aquellos que se hayan construido al
 considerar las reglas de inferencia.
 Particularmente estaremos interesados en el estudio de la lógica simbólica, que hace
 uso de símbolos para representar tanto los hechos como las reglas de inferencia.

### **Conceptos Clave:**  
- **Proposiciones:** Enunciados con valor de verdad (ej: $p \equiv$ "2 es par").  
- **Predicados:** Funciones lógicas con variables (ej: $P(x) \equiv$ "$x$ es primo").  
- **Registros Semióticos (Duval):**  
  - **Simbólico:** Notación formal ($\forall x \in \mathbb{N}, P(x)$).  
  - **Gráfico:** Diagramas en GeoGebra.  
  - **Concreto:** Circuitos en Crocodile Clips.

## **1. Concepto de Lógica Difusa**
En la lógica difusa, los valores de verdad no son solo `0` (falso) o `1` (verdadero), sino cualquier número en el intervalo `[0, 1]`. Esto permite modelar situaciones donde las categorías no son absolutas (ej: "caliente", "tibio", "frío").

````python

import numpy as np # crear arreglos numéricos como vectores  o matrices. Tablas. 
import skfuzzy as fuzz # para el manejo de datos difusos 
from skfuzzy import control as ctrl # para el manejo de datos difusos y el control de sistemas
import matplotlib.pyplot as plt # para graficar

# 1. Definir variables difusas
temperatura = ctrl.Antecedent(np.arange(0, 41, 1), 'temperatura')
velocidad = ctrl.Consequent(np.arange(0, 101, 1), 'velocidad')

# 2. Funciones de membresía (grados de verdad)
temperatura['fría'] = fuzz.trapmf(temperatura.universe, [0, 0, 10, 20])
temperatura['tibia'] = fuzz.trimf(temperatura.universe, [10, 20, 30])
temperatura['caliente'] = fuzz.trapmf(temperatura.universe, [20, 30, 40, 40])

velocidad['baja'] = fuzz.trimf(velocidad.universe, [0, 0, 50])
velocidad['media'] = fuzz.trimf(velocidad.universe, [0, 50, 100])
velocidad['alta'] = fuzz.trimf(velocidad.universe, [50, 100, 100])

# 3. Visualizar funciones
temperatura.view()
velocidad.view()
plt.show()

````


## **3. Reglas Difusas y Simulación**


````python

# 4. Reglas difusas (ej: "Si temperatura es caliente, entonces velocidad es alta")
regla1 = ctrl.Rule(temperatura['fría'], velocidad['baja'])
regla2 = ctrl.Rule(temperatura['tibia'], velocidad['media'])
regla3 = ctrl.Rule(temperatura['caliente'], velocidad['alta'])

# 5. Sistema de control
sistema_control = ctrl.ControlSystem([regla1, regla2, regla3])
controlador = ctrl.ControlSystemSimulation(sistema_control)

# 6. Simular para una temperatura de 25°C
controlador.input['temperatura'] = 25
controlador.compute()

# 7. Resultado
print(f"Velocidad del ventilador: {controlador.output['velocidad']:.2f}%")
velocidad.view(sim=controlador)
plt.show()

````

````python
import numpy as np # para cálculo numérico 
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Semáforo inteligente: lógica predicativa en Python

import matplotlib.pyplot as plt

# Definimos el predicado: "Es seguro cruzar" (S) en función de sensores
def S(vehiculos, peaton):
    # Si hay vehículos y hay peatón esperando, no es seguro cruzar
    # Si no hay vehículos y hay peatón, es seguro cruzar
    # Si hay vehículos y no hay peatón, luz verde para autos
    # Si no hay vehículos ni peatón, luz en rojo (ahorro)
    if peaton and not vehiculos:
        return 'Verde peatón'
    elif vehiculos and not peaton:
        return 'Verde autos'
    elif peaton and vehiculos:
        return 'Rojo peatón'
    else:
        return 'Rojo total'

# Simulación de sensores en el tiempo
tiempo = np.arange(0, 20, 1)
vehiculos = np.random.choice([True, False], size=tiempo.shape)
peaton = np.random.choice([True, False], size=tiempo.shape)

estados = [S(v, p) for v, p in zip(vehiculos, peaton)]

# Visualización animada del semáforo
colores = {'Verde peatón': 'green', 'Verde autos': 'blue', 'Rojo peatón': 'red', 'Rojo total': 'gray'}

fig, ax = plt.subplots(figsize=(5, 3))
def animate(i):
    ax.clear()
    estado = estados[i]
    ax.set_title(f"t={i} - Estado: {estado}")
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    # Dibuja el semáforo
    if estado == 'Verde peatón':
        ax.add_patch(plt.Circle((0.5, 0.7), 0.1, color='green'))
        ax.add_patch(plt.Circle((0.5, 0.5), 0.1, color='gray'))
        ax.add_patch(plt.Circle((0.5, 0.3), 0.1, color='gray'))
    elif estado == 'Verde autos':
        ax.add_patch(plt.Circle((0.5, 0.7), 0.1, color='gray'))
        ax.add_patch(plt.Circle((0.5, 0.5), 0.1, color='gray'))
        ax.add_patch(plt.Circle((0.5, 0.3), 0.1, color='blue'))
    elif estado == 'Rojo peatón':
        ax.add_patch(plt.Circle((0.5, 0.7), 0.1, color='red'))
        ax.add_patch(plt.Circle((0.5, 0.5), 0.1, color='gray'))
        ax.add_patch(plt.Circle((0.5, 0.3), 0.1, color='gray'))
    else:
        ax.add_patch(plt.Circle((0.5, 0.7), 0.1, color='gray'))
        ax.add_patch(plt.Circle((0.5, 0.5), 0.1, color='gray'))
        ax.add_patch(plt.Circle((0.5, 0.3), 0.1, color='gray'))
    ax.axis('off')

anim = FuncAnimation(fig, animate, frames=len(tiempo), interval=800)
HTML(anim.to_jshtml())
````
