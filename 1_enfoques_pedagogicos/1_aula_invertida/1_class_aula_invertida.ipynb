```python
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde el archivo CSV
# path = 'C:/Users/marco/Documentos/docencia/groups_list/logica.csv'
# estudiantes = pd.read_csv(path)
# estudiantes['Asistencia'] = estudiantes['Asistencia'].astype(float)  # Asegurar que la columna 'Parcial' es float

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    global lista_estudiantes
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        # Validación de entrada
        if asistencia == 'n':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Asistencia'] += 4
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        elif asistencia != 's':
            print("Entrada inválida. Por favor, introduce 's' para sí o 'n' para no.")
            return  # Volver a preguntar para el mismo estudiante si la entrada es inválida
        
        # Remover al estudiante de la lista para no llamarlo de nuevo
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Función para guardar cambios al final de la sesión
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()

# Guardar los cambios una vez que todos los estudiantes hayan sido llamados
guardar_cambios()

```

**Código para Generar el Mapa:**  


```python
from graphviz import Digraph

dot = Digraph(comment='Clase Lógica Matemática', engine='dot')  # Usar "dot" para mejor jerarquización
dot.attr(graph='''  
    overlap=false;        # Desactiva superposición de nodos  
    splines=true;         # Bordes curvos para evitar cruces  
    nodesep=1.0;          # Espacio horizontal entre nodos  
    ranksep=1.5;          # Espacio vertical entre niveles  
    ratio=auto;           # Ajusta automáticamente la relación de aspecto  
    concentrate=false     # No fusionar bordes  
''')

# ========== NODOS PRINCIPALES ========== (Configurar forma y color)
dot.node('A', 'Aula Invertida\n(Videos en YouTube, GitHub)', 
         shape='box3d', color='blue', style='filled', fillcolor='#E6F3FF', fontname='Helvetica')
dot.node('B', 'STEAMS\n(Python, SymPy, Jupyter)', 
         shape='ellipse', color='red', style='filled', fillcolor='#FFE6E6', fontname='Helvetica')
dot.node('C', 'ABPP\n(Proyectos con IA y Circuitos)', 
         shape='component', color='green', style='filled', fillcolor='#E6FFE6', fontname='Helvetica')
dot.node('D', 'Lógica Matemática\n(Teoría de Conjuntos, Demostraciones)', 
         shape='note', color='purple', fontname='Helvetica')

# ========== SUBNODOS ========== (Agrupar bajo el nodo "D")
with dot.subgraph(name='cluster_D') as sub:  # Subgrafo para organización
    sub.attr(style='dashed', label='Herramientas', fontcolor='purple')
    sub.node('D1', 'Circuitos en Crocodile Clips', shape='folder', color='gray')
    sub.node('D2', 'Animaciones con Matplotlib', shape='triangle', color='orange')
    sub.node('D3', 'ChatGPT para Validar Demostraciones', shape='hexagon', color='#00CC99')

# ========== CONEXIONES ========== (Jerarquía clara)
dot.edges(['AB', 'AC', 'BC', 'AD', 'BD', 'CD'])  # Conexiones principales
dot.edge('D', 'D1', style='dashed')  # Bordes punteados para subnodos
dot.edge('D', 'D2', style='dashed')  
dot.edge('D', 'D3', style='dashed')  

# ========== RENDERIZAR ========== (Asegurar rutas válidas)
dot.render(directory=r"C:\Users\marco\Downloads",  
           filename='mapa_logica_mejorado',  
           format='png',  
           view=True,  
           cleanup=True)
```

- **Python/SymPy:**  


```python
# Ejemplo: Tabla de Verdad con SymPy  
from sympy import symbols, Equivalent  
p, q = symbols('p q')  
expr = Equivalent(p >> q, ~p | q)  # p implica q es equivalente a ~p o q
print(expr.simplify())  # Salida: True  
# implicación en sympy >>
# negación en sympy ~
# disyunción en sympy |
# En este scripts de python, nos muestra que sympy utiliza para las expresiones lógicas, notación aditiva. 
# Pero también Sympy tiene notación funcional Implies(), Not(), Or() para expresiones lógicas 
# en notación funcional.
```

- **Matplotlib Animations:**  


```python
import matplotlib.pyplot as plt 
from matplotlib.animation import FuncAnimation  
from IPython.display import HTML  

fig, ax = plt.subplots()  
x = np.linspace(0, 2*np.pi, 100)  
line, = ax.plot(x, np.sin(x))  

def animate(i):  
    line.set_ydata(np.sin(x + i/10))  
    return line,  

ani = FuncAnimation(fig, animate, frames=100, interval=50)  
HTML(ani.to_jshtml())  
  
```
