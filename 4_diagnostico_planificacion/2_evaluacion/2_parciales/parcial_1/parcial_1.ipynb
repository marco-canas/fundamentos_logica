### 7) CNF de \$(p \to q) \land (\neg q)\$ (15 pts)

1. Reescritura del condicional: $(p \to q) \equiv (\neg p \lor q)$.
2. Sustituci√≥n: $(\neg p \lor q) \land (\neg q)$.

**Esto ya est√° en CNF** (conjunci√≥n de cl√°usulas disyuntivas).

**Simplificaci√≥n adicional (opcional):** usando resoluci√≥n o absorci√≥n con la unidad \$\neg q\$:

* De \$(\neg p \lor q)\$ y \$\neg q\$ se deduce \$\neg p\$.

**Respuestas v√°lidas:**

* CNF simplificada: \$\neg p \land \neg q\$.

### 8) DNF de \$(p \lor (q \to r))\$ (15 pts)

1. Reescritura del condicional: \$(q \to r) \equiv (\neg q \lor r)\$.
2. Sustituci√≥n y asociaci√≥n con \$\lor\$ conmutativo:
   \$p \lor (\neg q \lor r) \equiv (p \lor \neg q \lor r)\$.

**Esto ya es DNF** (disyunci√≥n de *t√©rminos*, donde cada t√©rmino es una conjunci√≥n de literales; en este caso, t√©rminos unitarios).

**DNF can√≥nica (opcional, por minterminos):**
La f√≥rmula es verdadera salvo en el caso \$p=F, q=T, r=F\$. Por tanto, la DNF can√≥nica puede escribirse como la disyunci√≥n de todos los minterminos verdaderos:

\$(p\land q\land r) \lor (p\land q\land \neg r) \lor (p\land \neg q\land r) \lor (p\land \neg q\land \neg r) \lor (\neg p\land \neg q\land r) \lor (\neg p\land \neg q\land \neg r) \lor (\neg p\land q\land r)\$.

Ambas representaciones son equivalentes; para efectos de examen, **\$p \lor \neg q \lor r\$** es una DNF correcta y m√°s simple.

# Sobre las nociones de Forma Normal Conjuntiva y Forma Normal Disyuntiva  

En el **parcial anterior** te ped√≠ transformar proposiciones a **Forma Normal Conjuntiva (CNF)** y a **Forma Normal Disyuntiva (DNF)**. Te aclaro qu√© significa cada una, con ejemplos sencillos para que lo tengas claro al momento de retroalimentar a tus estudiantes:

---

## üîπ Forma Normal Conjuntiva (CNF)

* **Definici√≥n:** Una proposici√≥n est√° en **CNF** si se expresa como una **conjunci√≥n (AND, ‚àß)** de **cl√°usulas**, y cada cl√°usula es una **disyunci√≥n (OR, ‚à®)** de literales.
* **Literal:** una variable proposicional $p$ o su negaci√≥n $\neg p$.
* **Estructura general:**
  $(l_{11} \lor l_{12} \lor \dots \lor l_{1k}) \land (l_{21} \lor l_{22} \lor \dots \lor l_{2m}) \land \dots$
* **Ejemplo:**
  $(p \lor q) \land (\neg p \lor r) \land (q \lor \neg r)$
  Aqu√≠ cada par√©ntesis es una **cl√°usula disyuntiva**.
* **Idea intuitiva:** Es una lista de ‚Äúcondiciones‚Äù que deben cumplirse todas a la vez.

---

## üîπ Forma Normal Disyuntiva (DNF)

* **Definici√≥n:** Una proposici√≥n est√° en **DNF** si se expresa como una **disyunci√≥n (OR, ‚à®)** de **t√©rminos**, y cada t√©rmino es una **conjunci√≥n (AND, ‚àß)** de literales.
* **Literal:** igual que antes, \$p\$ o \$\neg p\$.
* **Estructura general:**
  $(l_{11} \land l_{12} \land \dots \land l_{1k}) \lor (l_{21} \land l_{22} \land \dots \land l_{2m}) \lor \dots$
* **Ejemplo:**
  $(p \land q) \lor (\neg p \land r) \lor (q \land \neg r)$
  Aqu√≠ cada par√©ntesis es un **t√©rmino conjuntivo**.
* **Idea intuitiva:** Es una lista de ‚Äúcasos posibles‚Äù que bastan para que la proposici√≥n sea verdadera.

---

## üîπ Diferencia esencial

* **CNF** = conjunci√≥n de disyunciones ‚Üí √∫til en **SAT solvers** y deducci√≥n l√≥gica.
* **DNF** = disyunci√≥n de conjunciones ‚Üí √∫til en **an√°lisis de casos** y clasificaci√≥n de contingencias.
