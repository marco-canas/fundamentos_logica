## Parte I. Sintaxis y Semántica (20 pts)

### 1) (20 puntos) Muestre que la proposición $[(p \to q) \land (q \to r)] \to (p \to r)$ describe una forma de razonar correcta, mostrando que su tabla de verdad es una tautología. Ejemplifique verbalmente esta proposición para darle sentido cotidiano, deportivo, económico o matemático. 

### 2) FBF (10 pts) (Opcional)

Determine si cada expresión es una *fórmula bien formada* (fbf).



**a)** $((p \land q) \lor r)$
**Solución:** **Sí es fbf.** Paréntesis correctos, conectivos binarios entre proposiciones bien formadas.



**b)** $(p \to )q$
**Solución:** **No es fbf.** Falta el consecuente dentro de los paréntesis del condicional; la sintaxis correcta sería $(p \to q)$.

**c)** $\neg(p \land (q \lor r))$
**Solución:** **Sí es fbf.** Negación unaria bien aplicada y paréntesis balanceados.

**d)** $(\lor p q)$
**Solución:** **No es fbf** en notación infija. El conectivo $\lor$ debe ir entre dos fórmulas, p. ej. $(p \lor q)$.

---



### 2) Traducción y evaluación (10 pts)

Sea \$p\$: “Hoy llueve”, \$q\$: “Traje paraguas”.

**i.** “Si hoy llueve, entonces traje paraguas.”
**Traducción:** \$p \to q\$
**Evaluación con \$p=F, q=V\$:** \$F \to V = V\$.

**ii.** “Hoy no llueve y traje paraguas.”
**Traducción:** \$\neg p \land q\$
**Evaluación con \$p=F, q=V\$:** \$\neg F \land V = T \land V = T\$.

## Parte II. Tablas de verdad (25 pts)

### 3) Tabla de \$(p \lor q) \to (\neg p \land q)\$ (15 pts)

| p | q | \$p\lor q\$ | \$\neg p\$ | \$\neg p \land q\$ | \$(p\lor q) \to (\neg p \land q)\$ |
| - | - | ----------: | ---------: | -----------------: | ---------------------------------: |
| T | T |           T |          F |                  F |                              **F** |
| T | F |           T |          F |                  F |                              **F** |
| F | T |           T |          T |                  T |                              **T** |
| F | F |           F |          T |                  F |                              **T** |

**Clasificación:** **Contingencia** (toma valores V y F).

---

### 4) Equivalencia \$p \to q \equiv \neg p \lor q\$ (10 pts)

| p | q | \$p\to q\$ | \$\neg p\$ | \$\neg p \lor q\$ |
| - | - | ---------: | ---------: | ----------------: |
| T | T |          T |          F |             **T** |
| T | F |          F |          F |             **F** |
| F | T |          T |          T |             **T** |
| F | F |          T |          T |             **T** |

Se observa igualdad columna a columna; por tanto, **\$p \to q\$ es lógicamente equivalente a \$\neg p \lor q\$**.

---

## Parte III. Equivalencias y Leyes (25 pts)

### 5) Simplificación de \$p \land (p \lor q)\$ (10 pts)

**Objetivo:** Simplificar usando leyes clásicas.

1. \$p \land (p \lor q)\$
2. **Absorción:** \$p \land (p \lor q) \equiv p\$.

> Alternativa detallada:
> \$p \land (p \lor q) \equiv (p \land p) \lor (p \land q)\$ (distributiva)
> \$\equiv p \lor (p \land q)\$ (idempotencia)
> \$\equiv p\$ (absorción).

**Resultado:** \$\boxed{p}\$.

---

### 6) Leyes de De Morgan (15 pts)

Demostrar \$\neg(p \lor q) \equiv (\neg p \land \neg q)\$.

**Demostración por leyes:**

1. \$\neg(p \lor q)\$
2. Por De Morgan (definición/teorema): \$\equiv (\neg p) \land (\neg q)\$.

**Verificación por tabla de verdad:**

| p | q | \$p\lor q\$ | \$\neg(p\lor q)\$ | \$\neg p\$ | \$\neg q\$ | \$(\neg p \land \neg q)\$ |
| - | - | ----------: | ----------------: | ---------: | ---------: | ------------------------: |
| T | T |           T |                 F |          F |          F |                     **F** |
| T | F |           T |                 F |          F |          T |                     **F** |
| F | T |           T |                 F |          T |          F |                     **F** |
| F | F |           F |                 T |          T |          T |                     **T** |

Las dos últimas columnas coinciden, por lo que la equivalencia es válida.

## Parte IV. Formas Normales (30 pts)

### 7) CNF de \$(p \to q) \land (\neg q)\$ (15 pts)

1. Reescritura del condicional: $(p \to q) \equiv (\neg p \lor q)$.
2. Sustitución: $(\neg p \lor q) \land (\neg q)$.

**Esto ya está en CNF** (conjunción de cláusulas disyuntivas).

**Simplificación adicional (opcional):** usando resolución o absorción con la unidad \$\neg q\$:

* De \$(\neg p \lor q)\$ y \$\neg q\$ se deduce \$\neg p\$.
* Equivalente a \$\neg p \land \neg q\$ (CNF con dos cláusulas unitarias).

**Respuestas válidas:**

* CNF sin simplificar: \$(\neg p \lor q) \land (\neg q)\$
* CNF simplificada: \$\neg p \land \neg q\$.

---

### 8) DNF de \$(p \lor (q \to r))\$ (15 pts)

1. Reescritura del condicional: \$(q \to r) \equiv (\neg q \lor r)\$.
2. Sustitución y asociación con \$\lor\$ conmutativo:
   \$p \lor (\neg q \lor r) \equiv (p \lor \neg q \lor r)\$.

**Esto ya es DNF** (disyunción de *términos*, donde cada término es una conjunción de literales; en este caso, términos unitarios).

**DNF canónica (opcional, por minterminos):**
La fórmula es verdadera salvo en el caso \$p=F, q=T, r=F\$. Por tanto, la DNF canónica puede escribirse como la disyunción de todos los minterminos verdaderos:

\$(p\land q\land r) \lor (p\land q\land \neg r) \lor (p\land \neg q\land r) \lor (p\land \neg q\land \neg r) \lor (\neg p\land \neg q\land r) \lor (\neg p\land \neg q\land \neg r) \lor (\neg p\land q\land r)\$.

Ambas representaciones son equivalentes; para efectos de examen, **\$p \lor \neg q \lor r\$** es una DNF correcta y más simple.

# Sobre las nociones de Forma Normal Conjuntiva y Forma Normal Disyuntiva  

En el **parcial anterior** te pedí transformar proposiciones a **Forma Normal Conjuntiva (CNF)** y a **Forma Normal Disyuntiva (DNF)**. Te aclaro qué significa cada una, con ejemplos sencillos para que lo tengas claro al momento de retroalimentar a tus estudiantes:

---

## 🔹 Forma Normal Conjuntiva (CNF)

* **Definición:** Una proposición está en **CNF** si se expresa como una **conjunción (AND, ∧)** de **cláusulas**, y cada cláusula es una **disyunción (OR, ∨)** de literales.
* **Literal:** una variable proposicional $p$ o su negación $\neg p$.
* **Estructura general:**
  $(l_{11} \lor l_{12} \lor \dots \lor l_{1k}) \land (l_{21} \lor l_{22} \lor \dots \lor l_{2m}) \land \dots$
* **Ejemplo:**
  $(p \lor q) \land (\neg p \lor r) \land (q \lor \neg r)$
  Aquí cada paréntesis es una **cláusula disyuntiva**.
* **Idea intuitiva:** Es una lista de “condiciones” que deben cumplirse todas a la vez.

---

## 🔹 Forma Normal Disyuntiva (DNF)

* **Definición:** Una proposición está en **DNF** si se expresa como una **disyunción (OR, ∨)** de **términos**, y cada término es una **conjunción (AND, ∧)** de literales.
* **Literal:** igual que antes, \$p\$ o \$\neg p\$.
* **Estructura general:**
  $(l_{11} \land l_{12} \land \dots \land l_{1k}) \lor (l_{21} \land l_{22} \land \dots \land l_{2m}) \lor \dots$
* **Ejemplo:**
  $(p \land q) \lor (\neg p \land r) \lor (q \land \neg r)$
  Aquí cada paréntesis es un **término conjuntivo**.
* **Idea intuitiva:** Es una lista de “casos posibles” que bastan para que la proposición sea verdadera.

---

## 🔹 Diferencia esencial

* **CNF** = conjunción de disyunciones → útil en **SAT solvers** y deducción lógica.
* **DNF** = disyunción de conjunciones → útil en **análisis de casos** y clasificación de contingencias.
