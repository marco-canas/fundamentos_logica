### 5) SimplificaciÃ³n de \$p \land (p \lor q)\$ (10 pts)

## Parte IV. Formas Normales (30 pts)

### 7) CNF de \$(p \to q) \land (\neg q)\$ (15 pts)

1. Reescritura del condicional: $(p \to q) \equiv (\neg p \lor q)$.
2. SustituciÃ³n: $(\neg p \lor q) \land (\neg q)$.

**Esto ya estÃ¡ en CNF** (conjunciÃ³n de clÃ¡usulas disyuntivas).

**SimplificaciÃ³n adicional (opcional):** usando resoluciÃ³n o absorciÃ³n con la unidad \$\neg q\$:

* De \$(\neg p \lor q)\$ y \$\neg q\$ se deduce \$\neg p\$.
* Equivalente a \$\neg p \land \neg q\$ (CNF con dos clÃ¡usulas unitarias).

**Respuestas vÃ¡lidas:**

* CNF sin simplificar: \$(\neg p \lor q) \land (\neg q)\$
* CNF simplificada: \$\neg p \land \neg q\$.

### 8) DNF de \$(p \lor (q \to r))\$ (15 pts)

1. Reescritura del condicional: \$(q \to r) \equiv (\neg q \lor r)\$.
2. SustituciÃ³n y asociaciÃ³n con \$\lor\$ conmutativo:
   \$p \lor (\neg q \lor r) \equiv (p \lor \neg q \lor r)\$.

**Esto ya es DNF** (disyunciÃ³n de *tÃ©rminos*, donde cada tÃ©rmino es una conjunciÃ³n de literales; en este caso, tÃ©rminos unitarios).

**DNF canÃ³nica (opcional, por minterminos):**
La fÃ³rmula es verdadera salvo en el caso \$p=F, q=T, r=F\$. Por tanto, la DNF canÃ³nica puede escribirse como la disyunciÃ³n de todos los minterminos verdaderos:

\$(p\land q\land r) \lor (p\land q\land \neg r) \lor (p\land \neg q\land r) \lor (p\land \neg q\land \neg r) \lor (\neg p\land \neg q\land r) \lor (\neg p\land \neg q\land \neg r) \lor (\neg p\land q\land r)\$.

Ambas representaciones son equivalentes; para efectos de examen, **\$p \lor \neg q \lor r\$** es una DNF correcta y mÃ¡s simple.

# Sobre las nociones de Forma Normal Conjuntiva y Forma Normal Disyuntiva  

En el **parcial anterior** te pedÃ­ transformar proposiciones a **Forma Normal Conjuntiva (CNF)** y a **Forma Normal Disyuntiva (DNF)**. Te aclaro quÃ© significa cada una, con ejemplos sencillos para que lo tengas claro al momento de retroalimentar a tus estudiantes:

---

## ğŸ”¹ Forma Normal Conjuntiva (CNF)

* **DefiniciÃ³n:** Una proposiciÃ³n estÃ¡ en **CNF** si se expresa como una **conjunciÃ³n (AND, âˆ§)** de **clÃ¡usulas**, y cada clÃ¡usula es una **disyunciÃ³n (OR, âˆ¨)** de literales.
* **Literal:** una variable proposicional $p$ o su negaciÃ³n $\neg p$.
* **Estructura general:**
  $(l_{11} \lor l_{12} \lor \dots \lor l_{1k}) \land (l_{21} \lor l_{22} \lor \dots \lor l_{2m}) \land \dots$
* **Ejemplo:**
  $(p \lor q) \land (\neg p \lor r) \land (q \lor \neg r)$
  AquÃ­ cada parÃ©ntesis es una **clÃ¡usula disyuntiva**.
* **Idea intuitiva:** Es una lista de â€œcondicionesâ€ que deben cumplirse todas a la vez.

---

## ğŸ”¹ Forma Normal Disyuntiva (DNF)

* **DefiniciÃ³n:** Una proposiciÃ³n estÃ¡ en **DNF** si se expresa como una **disyunciÃ³n (OR, âˆ¨)** de **tÃ©rminos**, y cada tÃ©rmino es una **conjunciÃ³n (AND, âˆ§)** de literales.
* **Literal:** igual que antes, \$p\$ o \$\neg p\$.
* **Estructura general:**
  $(l_{11} \land l_{12} \land \dots \land l_{1k}) \lor (l_{21} \land l_{22} \land \dots \land l_{2m}) \lor \dots$
* **Ejemplo:**
  $(p \land q) \lor (\neg p \land r) \lor (q \land \neg r)$
  AquÃ­ cada parÃ©ntesis es un **tÃ©rmino conjuntivo**.
* **Idea intuitiva:** Es una lista de â€œcasos posiblesâ€ que bastan para que la proposiciÃ³n sea verdadera.

---

## ğŸ”¹ Diferencia esencial

* **CNF** = conjunciÃ³n de disyunciones â†’ Ãºtil en **SAT solvers** y deducciÃ³n lÃ³gica.
* **DNF** = disyunciÃ³n de conjunciones â†’ Ãºtil en **anÃ¡lisis de casos** y clasificaciÃ³n de contingencias.
