Muestre que la siguiente proposición es una tautología utilizando solo reglas o leyes de la lógica  proposicional: 

$$
[(p \lor q) \land \neg p] \to q
$$




$$
\begin{aligned}
[(p \lor q) \land \neg p] \to q 
&\equiv \neg[(p \lor q) \land \neg p] \lor q \quad \text{(Definición alterna del condicional)} \\
&\equiv [\neg(p \lor q) \lor \neg\neg p] \lor q \quad \text{(De Morgan)} \\
&\equiv [(\neg p \land \neg q) \lor p] \lor q \quad \text{(Ley de la doble negación)} \\
&\equiv [(\neg p \lor p) \land (\neg q \lor p)] \lor q \quad \text{(Distributividad)} \\
&\equiv [V \land (\neg q \lor p)] \lor q \quad \text{(Ley del tercero excluido)} \\
&\equiv (\neg q \lor p) \lor q \quad \text{(Ley de identidad)} \\
&\equiv \neg q \lor (p \lor q) \quad \text{(Ley asociativa)} \\
&\equiv \neg q \lor (q \lor p) \quad \text{(Ley conmutativa)} \\
&\equiv (\neg q \lor q) \lor p \quad \text{(Ley asociativa)} \\
&\equiv V \lor p \quad \text{(Ley del tercero excluido)} \\
&\equiv V \quad \text{(Ley de identidad)}
\end{aligned}
$$

Tabla de verdad para determinar si un argummento es válido o no válido.  

Premisas: 
* $p$ 
* $p \to q$
* $\neg q \lor r$

Conclusión: $r$

## Solución:  



Argumentación como proposición compuesta o representación horizontal: 

$$
[p \land (p \to q) \land (\neg q \lor r)] \to r
$$

```python
valores_de_verdad = [True, False] 

for p in valores_de_verdad:
    for q in valores_de_verdad:
        for r in valores_de_verdad:
            izquierda = p and (p <= q) and (not q or r)
            implicacion = (not izquierda) or r
            print(f"p: {p}, q: {q}, r: {r} => (p ∧ (p → q) ∧ (¬q ∨ r)) → r: {implicacion}")
```
