```python
"""
Generador de .docx para Quiz 1 - Fundamentos de L√≥gica
Guarda: quiz1_fundamentos_logica.docx
Requisitos: pip install python-docx
"""

from docx import Document
from docx.shared import Pt
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT

def add_heading(document, text, level=1):
    h = document.add_heading(level=level)
    run = h.add_run(text)
    run.bold = True
    return h

def add_paragraph(document, text, bold=False, italic=False, size=11, align='left'):
    p = document.add_paragraph()
    run = p.add_run(text)
    run.bold = bold
    run.italic = italic
    run.font.size = Pt(size)
    if align == 'center':
        p.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    return p

def main():
    doc = Document()

    # Estilos generales (t√≠tulos)
    title = "Quiz 1 ‚Äî Fundamentos de L√≥gica"
    subtitle = "Licenciatura en Matem√°ticas ‚Äî Facultad de Educaci√≥n (Campus Caucasia)"
    add_heading(doc, title, level=1)
    add_paragraph(doc, subtitle, size=10, italic=True, align='center')
    add_paragraph(doc, "")  # l√≠nea en blanco

    # Metadata
    add_paragraph(doc, "Duraci√≥n: 30 minutos", bold=False)
    add_paragraph(doc, "Fecha: Lunes 11 de agosto de 2025", bold=False)
    add_paragraph(doc, "Valor: 5% de la nota final", bold=False)
    add_paragraph(doc, "Instrucciones:", bold=True)
    add_paragraph(doc, "1. Responda con letra legible.")
    add_paragraph(doc, "2. Justifique cada respuesta con los procedimientos solicitados.")
    add_paragraph(doc, "3. No se permite el uso de celulares. Puede usar calculadora cient√≠fica (Casio 570 o similar).")
    add_paragraph(doc, "")  # l√≠nea en blanco

    # Preguntas
    add_paragraph(doc, "Preguntas:", bold=True)

    # Pregunta 1
    q1_title = "1. (2 puntos) Clasifique las siguientes expresiones en proposiciones o no proposiciones, y justifique:"
    add_paragraph(doc, q1_title, bold=True)
    add_paragraph(doc, "a) 2 + 3 = 5")
    add_paragraph(doc, "b) ¬øQu√© hora es?")
    add_paragraph(doc, "c) Si hoy llueve, entonces llevo sombrilla.")
    add_paragraph(doc, "d) El n√∫mero 7 es par.")
    add_paragraph(doc, "")  # espacio

    # Pregunta 2
    q2_title = "2. (2 puntos) Construya la tabla de verdad de la proposici√≥n compuesta:"
    add_paragraph(doc, q2_title, bold=True)
    # Mostrar la f√≥rmula con Unicode
    formula_q2 = "(p ‚àß ¬¨q) ‚à® r"
    add_paragraph(doc, f"F√≥rmula: {formula_q2}")
    add_paragraph(doc, "Construya la tabla de verdad completa (liste todas las combinaciones de p, q, r).")
    add_paragraph(doc, "")  # espacio

    # Pregunta 3
    q3_title = "3. (2 puntos)"
    add_paragraph(doc, q3_title, bold=True)
    add_paragraph(doc, "Dada la proposici√≥n:")
    add_paragraph(doc, "(p ‚Üí q) ‚àß (q ‚Üí r)")
    add_paragraph(doc, "a) Interprete su significado en lenguaje natural.")
    add_paragraph(doc, "b) Eval√∫e su valor de verdad cuando p = V, q = F, r = V.")
    add_paragraph(doc, "")  # espacio

    # Pregunta 4
    q4_title = "4. (2 puntos) Determine si las proposiciones siguientes son equivalentes. Justifique usando tablas de verdad:"
    add_paragraph(doc, q4_title, bold=True)
    add_paragraph(doc, "a) p ‚Üí q")
    add_paragraph(doc, "b) ¬¨p ‚à® q")
    add_paragraph(doc, "")  # espacio

    # Pregunta 5
    q5_title = "5. (2 puntos) Formule en lenguaje simb√≥lico usando:"
    add_paragraph(doc, q5_title, bold=True)
    add_paragraph(doc, "- p: \"Hoy es lunes\"")
    add_paragraph(doc, "- q: \"Tenemos clase de Fundamentos de L√≥gica\"")
    add_paragraph(doc, "- r: \"Hay examen\"")
    add_paragraph(doc, "La frase: \"Si hoy es lunes y tenemos clase, entonces no hay examen.\"")
    add_paragraph(doc, "")
    add_paragraph(doc, "-----------------------------------------------")
    add_paragraph(doc, "¬°Buena suerte!", align='center')

    # Guardar archivo
    filename = "quiz1_fundamentos_logica.docx"
    doc.save(filename)
    print(f"Archivo guardado: {filename}")

if __name__ == "__main__":
    main()
```
