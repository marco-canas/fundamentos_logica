```python
# funci√≥n de usuario para el c√≥digo QR 

# Instalaci√≥n necesaria si no tienes qrcode
# # !pip install qrcode[pil]

import qrcode
from IPython.display import Image, display
from io import BytesIO

def generar_qr(url: str, box_size: int = 10, border: int = 4):
    """
    Genera y muestra en Jupyter un c√≥digo QR para la URL dada.
    
    Par√°metros:
    -----------
    url : str
        Direcci√≥n web o texto a codificar en el QR.
    box_size : int
        Tama√±o de cada caja (pixel) en el QR.
    border : int
        Grosor del borde del QR (en cajas).
    """
    # Crear el objeto QR
    qr = qrcode.QRCode(
        version=1,
        box_size=box_size,
        border=border
    )
    qr.add_data(url)
    qr.make(fit=True)

    # Convertir a imagen
    img = qr.make_image(fill_color="black", back_color="white")

    # Guardar en memoria para mostrar en Jupyter
    buffer = BytesIO()
    img.save(buffer, format="PNG")
    buffer.seek(0)
    display(Image(data=buffer.getvalue()))
```

# El operador l√≥gico XOR en t√©rminos de los operadores negaci√≥n, disyunci√≥n y conjunci√≥n  



El operador l√≥gico **xor** (‚Äúo exclusivo‚Äù) puede expresarse usando solo **disyunciones (‚à®)**, **conjunciones (‚àß)** y **negaciones (¬¨)**.

La definici√≥n m√°s com√∫n es:

$$
p \oplus q \equiv (p \lor q) \land \neg(p \land q)
$$



### Explicaci√≥n intuitiva

* **$p \lor q$** asegura que *al menos una* es verdadera.
* **$\neg(p \land q)$** asegura que *no ambas* son verdaderas.
* Juntas, expresan ‚Äúuna y solo una es verdadera‚Äù.

---

Tambi√©n puedes expresarlo de otras formas equivalentes:



# Forma usando dos conjunciones

$$
p \oplus q \equiv (p \land \neg q)\ \lor\ (\neg p \land q)
$$

```python
# Tabla de verdad del operador l√≥gico XOR sin SymPy

# Definici√≥n de XOR usando operadores de Python
def xor(p, q):
    return (p and not q) or (not p and q)
```

```python

# Valores posibles
valores = [False, True]
```

```python
print("p\tq\tp XOR q")
# Generar tabla
for p in valores:
    for q in valores:
        print(f"{p}\t{q}\t{xor(p, q)}")

```
