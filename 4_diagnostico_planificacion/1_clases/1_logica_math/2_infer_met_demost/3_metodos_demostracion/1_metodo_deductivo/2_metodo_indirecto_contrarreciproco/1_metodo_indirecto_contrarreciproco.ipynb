## 1. TÃ­tulo de la Clase: **"El poder del reverso: Demostrando implicaciones mediante el contrarrecÃ­proco"**



### IntroducciÃ³n  

En este mÃ©todo buscamos probar una implicaciÃ³n $H \to T$ pero la probamos de manera indirecta, es decir, la que probamos a a su implicaciÃ³n equivalente 

$$ \neg T \to \neg H $$

Recordemos todo lo que hemos visto con el contrarrecÃ­proco. 

1. Las cuatro implicaciones asiciadas: $p \to q$, $q \to p$, $\neg p \to \neg q$, y $\neg q \to \neg p$. 
2. Modus Tollendo Tollens $[(p \to q) \land \neg q] \to \neg p$ 

## Recordar 

* ImplicaciÃ³n: $p \to q$ significa que si $p$ es verdadero, entonces $q$ tambiÃ©n es verdadero.
* $q \to p$: esta es el recÃ­proco de la implicaciÃ³n anterior 
* $\neg p \to \neg q$: la implicaciÃ³n "contraria" de la implicaciÃ³n original $p \to q$. 
* $\neg q \to \neg p $: esta es la contrarrecÃ­proca o contrapositiva de la implicaciÃ³n original $p \to q$.

### Argumentos por contraposiciÃ³n   

 Una segunda forma de argumentaciÃ³n indirecta, *argumento por contraposiciÃ³n*, se basa en 
la equivalencia lÃ³gica entre un enunciado y su contrapositivo. 

Para probar un enunciado por contraposiciÃ³n, se toma el contrapositivo del enunciado, se demuestra el contrapositivo en una demostraciÃ³n directa y se concluye que el enunciado original es verdadero. 

El razonamiento  subyacente es que, dado que un enunciado condicional es lÃ³gicamente equivalente a su contra
positivo, si el contrapositivo es verdadero, entonces el enunciado tambiÃ©n debe ser verdadero.

# Uso de la notaciÃ³n de cuantificadores 

$$ (\text{Contextos})(\text{FunciÃ³n proposicional}) $$

$$ (\forall n \in \mathbb{N})(n^{2} \quad \text{impar} \implies n \quad \text{impar} ) $$

    b. Demuestre que $P(x)$ es falso

### Ejemplo 4.6.4 Si el cuadrado de un nÃºmero entero es par, entonces el entero es par  



**Demuestre que para todo entero $n$, si $n^{2}$ es par entonces $n$ es par.**   



**SoluciÃ³n** 

Primero forma que el contrapositivo del enunciado a ser probado.  


Contrapositivo: Para todo entero $n$, si $n$ no es par entonces $n^{2}$ no es par.


Por el teorema del cociente-residuo con $d à¸ 2$, cualquier nÃºmero entero es par o impar, por lo que cualquier nmero entero que no es par es impar. TambiÃ©n por el teorema 4.6.2, ningn entero puede ser a la vez par e impar. AsÃ­ que si un nmero entero es impar, entonces no es par. 

AsÃ­, el contrapositivo se puede replantear de la siguiente manera:  

Contrapositivo: Para todos los enteros $n$, si $n$ es impar, entonces $n^{2}$ es impar.  

Un cÃ¡lculo sencillo es el corazÃ³n de una demostraciÃ³n directa de este enunciado, como se muestra a continuaciÃ³n  



## ProposiciÃ³n 4.6.4
 Para todos los enteros $n$, si $n^{2}$ es par entonces $n$ es par.  



 **DemostraciÃ³n (por contraposiciÃ³n):**  
 
Supongamos que $n$ es un entero impar. [Debemos demostrar que $n^{2}$ es impar.] Por definiciÃ³n de impar, n à¸ 2k à¸ 1 para algn entero k. Por sustituci n y lgebra, n2 à¸ (2k à¸ 1)2 à¸ 4k 2 à¸ 4k à¸ 1 à¸ 2(2k 2 à¸ 2k) à¸ 1.


Pero 2k2 à¸ 2k es un nmero entero, porque los productos y las sumas de los nmeros enteros son nmeros enteros. AsÃ­ n2 à¸ 2 à¸ (un entero) à¸ 1 y por tanto, por definici n, de impar, n2 es impar [que era lo que se querÃ­a demostrar].


## Ejemplo MatemÃ¡tico:  



**Teorema**: "Si $n^2$ es impar, entonces $n$ es impar"


````python
# VerificaciÃ³n de este teorema con Python 
from sympy import symbols, Implies, Not, simplify_logic # CÃ¡lculo simbÃ³lico con SymPy

n = symbols('n', integer=True, nonnegative=True) # define la variable y ell contexto numÃ©rico 
n = 2
P = n**2 % 2 == 1 # P: n**2 es impar
Q = n % 2 == 1 # Q: n es impar

# ContrarrecÃ­proco
implicacion = Implies(P, Q)  # P implica Q
contrapositivo = Implies(Not(Q), Not(P))
print(n, n**2, P, Q, implicacion, contrapositivo)

````

### 7.3 En Python/Jupyter Notebook   

Esta es una VisualizaciÃ³n interactiva de la implicaciÃ³n y su contrarrecÃ­proco



````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import ipywidgets as widgets
from sympy import symbols, Implies, Not

# --- Parte 1: Widget Interactivo (Equivalencia LÃ³gica) ---
P, Q = symbols('P Q')

def mostrar_equivalencia(P_val, Q_val):
    # Evaluar la implicaciÃ³n directa P â Q
    implicacion_directa = Implies(P, Q).subs({P: P_val, Q: Q_val})
    
    # Evaluar el contrarrecÃ­proco Â¬Q â Â¬P
    contrarreciproco = Implies(Not(Q), Not(P)).subs({P: P_val, Q: Q_val})
    
    # Mostrar resultados
    print(f"ImplicaciÃ³n directa (P â Q): {implicacion_directa}")
    print(f"ContrarrecÃ­proco (Â¬Q â Â¬P): {contrarreciproco}")
    print(f"Â¿Son equivalentes? {implicacion_directa == contrarreciproco}")

# Crear widget interactivo
widgets.interact(
    mostrar_equivalencia,
    P_val=widgets.Dropdown(options=[True, False], description='P:'),
    Q_val=widgets.Dropdown(options=[True, False], description='Q:')
)

# --- Parte 2: AnimaciÃ³n (VisualizaciÃ³n del ContrarrecÃ­proco) ---
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_title("DemostraciÃ³n por ContrarrecÃ­proco")
ax.grid(True)

# Puntos para P, Q, Â¬P, Â¬Q
P_point, = ax.plot([], [], 'bo', markersize=15, label='P')
Q_point, = ax.plot([], [], 'go', markersize=15, label='Q')
notP_point, = ax.plot([], [], 'ro', markersize=15, label='Â¬P')
notQ_point, = ax.plot([], [], 'yo', markersize=15, label='Â¬Q')

# Flechas para las implicaciones
arrow_directa = ax.annotate('', xy=(0, 0), xytext=(0, 0), 
                           arrowprops=dict(arrowstyle='->', color='blue', lw=2))
arrow_contrar = ax.annotate('', xy=(0, 0), xytext=(0, 0), 
                           arrowprops=dict(arrowstyle='->', color='red', lw=2))

# Texto para las etiquetas
ax.text(2, 8, 'P', fontsize=12, ha='center', va='center')
ax.text(8, 8, 'Q', fontsize=12, ha='center', va='center')
ax.text(2, 2, 'Â¬P', fontsize=12, ha='center', va='center')
ax.text(8, 2, 'Â¬Q', fontsize=12, ha='center', va='center')

def init():
    # Posiciones iniciales de los puntos
    P_point.set_data([2], [8])
    Q_point.set_data([8], [8])
    notP_point.set_data([2], [2])
    notQ_point.set_data([8], [2])
    return P_point, Q_point, notP_point, notQ_point

def animate(i):
    # Dibujar flechas en diferentes pasos de la animaciÃ³n
    if i == 10:
        arrow_directa.set_position((2.2, 8))
        arrow_directa.xy = (7.8, 8)
    elif i == 30:
        arrow_contrar.set_position((7.8, 2))
        arrow_contrar.xy = (2.2, 2)
    return P_point, Q_point, notP_point, notQ_point, arrow_directa, arrow_contrar

# Crear y mostrar la animaciÃ³n
anim = FuncAnimation(fig, animate, init_func=init, frames=50, interval=200, blit=True)
plt.legend(loc='upper right')
plt.close()  # Evita que se muestre el grÃ¡fico estÃ¡tico antes de la animaciÃ³n
HTML(anim.to_jshtml())
````
