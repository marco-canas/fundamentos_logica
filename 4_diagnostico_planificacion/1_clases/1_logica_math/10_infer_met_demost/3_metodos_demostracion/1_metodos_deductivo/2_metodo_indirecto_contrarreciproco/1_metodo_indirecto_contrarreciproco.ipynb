# **"El poder del reverso: Demostrando implicaciones mediante el contrarrecíproco"**



### Introducción  

En este método buscamos probar una implicación $H \to T$ pero la probamos de manera indirecta, es decir, la que probamos a a su implicación equivalente 

$$ \neg T \to \neg H $$

Recordemos todo lo que hemos visto con el contrarrecíproco. 

1. Las cuatro implicaciones asiciadas: $p \to q$, $q \to p$, $\neg p \to \neg q$, y $\neg q \to \neg p$. 
2. Modus Tollendo Tollens $[(p \to q) \land \neg q] \to \neg p$ 

## Recordar 

* Implicación: $p \to q$ significa que si $p$ es verdadero, entonces $q$ también es verdadero.
* $q \to p$: esta es el recíproco de la implicación anterior 
* $\neg p \to \neg q$: la implicación "contraria" de la implicación original $p \to q$. 
* $\neg q \to \neg p $: esta es la contrarrecíproca o contrapositiva de la implicación original $p \to q$.

### Argumentos por contraposición   

 Una segunda forma de argumentación indirecta, *argumento por contraposición*, se basa en 
la equivalencia lógica entre un enunciado y su contrapositivo. 

Para probar un enunciado por contraposición, se toma el contrapositivo del enunciado, se demuestra el contrapositivo en una demostración directa y se concluye que el enunciado original es verdadero. 

El razonamiento  subyacente es que, dado que un enunciado condicional es lógicamente equivalente a su contra
positivo, si el contrapositivo es verdadero, entonces el enunciado también debe ser verdadero.

# Uso de la notación de cuantificadores 

$$ (\text{Contextos})(\text{Función proposicional}) $$

$$ (\forall n \in \mathbb{N})(n^{2} \quad \text{impar} \implies n \quad \text{impar} ) $$

# Método de demostración por contraposición  

 1. Exprese el enunciado a demostrar en la forma
 $$ \forall x \ \text{en} \ D, \ \text{si} \ P(x), \ \text{entonces} \ Q(x). $$
   (Este paso se puede hacer mentalmente.)
 2. Reescriba este enunciado en forma contrapositiva
 $$ \forall x \ \text{en} \ D, \ \text{si} \ Q(x) \ \text{es falso, entonces} \ P(x) \ \text{es falso.} $$
   (Este paso también se puede hacer mentalmente.)
 3. Demuestre el enunciado contrapositivo con una demostración directa,  
    a. Supongamos que $x$ es un elemento (particular, pero elegido arbitrariamente) de $D$ tal que $Q(x)$ es falso.  
    b. Demuestre que $P(x)$ es falso

### Ejemplo 4.6.4 Si el cuadrado de un número entero es par, entonces el entero es par  



**Demuestre que para todo entero $n$, si $n^{2}$ es par entonces $n$ es par.**   



**Solución** 

Primero forma que el contrapositivo del enunciado a ser probado.  


Contrapositivo: Para todo entero $n$, si $n$ no es par entonces $n^{2}$ no es par.


Por el teorema del cociente-residuo con $d = 2$, cualquier número entero es par o impar, por lo que cualquier nmero entero que no es par es impar. 

También por el teorema 4.6.2, ningn entero puede ser a la vez par e impar. 

Así que si un nmero entero es impar, entonces no es par. 

Así, el contrapositivo se puede replantear de la siguiente manera:  

Contrapositivo: Para todos los enteros $n$, si $n$ es impar, entonces $n^{2}$ es impar.  

Un cálculo sencillo es el corazón de una demostración directa de este enunciado, como se muestra a continuación  



## Proposición 4.6.4
 Para todos los enteros $n$, si $n^{2}$ es par entonces $n$ es par.  



 **Demostración (por contraposición):**  
 
Supongamos que $n$ es un entero impar. [Debemos demostrar que $n^{2}$ es impar.] 

Por definición de impar, $n = 2k + 1$ para algún entero $k$. 

Por sustitución y álgebra,   

$$
n^{2} = (2k + 1)^{2} = 4k^{2} + 4k + 1  = 2(2k^2 + 2k) + 1.
$$


Pero $2k^2 + 2k$ es un número entero, porque los productos y las sumas de los números enteros son números enteros. 

Así $n^{2} = 2 \cdot (\text{un entero}) + 1$ y por tanto, por definición, de impar, $n^2$ es impar [que era lo que se quería demostrar].

## Conclusión o reflexión

Veo que cada vez que elijo un número cuadrado impar, el tendrá su raíz también impar.

### 7.3 En Python/Jupyter Notebook   

Esta es una Visualización interactiva de la implicación y su contrarrecíproco



````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import ipywidgets as widgets
from sympy import symbols, Implies, Not

# --- Parte 1: Widget Interactivo (Equivalencia Lógica) ---
P, Q = symbols('P Q')

def mostrar_equivalencia(P_val, Q_val):
    # Evaluar la implicación directa P → Q
    implicacion_directa = Implies(P, Q).subs({P: P_val, Q: Q_val})
    
    # Evaluar el contrarrecíproco ¬Q → ¬P
    contrarreciproco = Implies(Not(Q), Not(P)).subs({P: P_val, Q: Q_val})
    
    # Mostrar resultados
    print(f"Implicación directa (P → Q): {implicacion_directa}")
    print(f"Contrarrecíproco (¬Q → ¬P): {contrarreciproco}")
    print(f"¿Son equivalentes? {implicacion_directa == contrarreciproco}")

# Crear widget interactivo
widgets.interact(
    mostrar_equivalencia,
    P_val=widgets.Dropdown(options=[True, False], description='P:'),
    Q_val=widgets.Dropdown(options=[True, False], description='Q:')
)

# --- Parte 2: Animación (Visualización del Contrarrecíproco) ---
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_title("Demostración por Contrarrecíproco")
ax.grid(True)

# Puntos para P, Q, ¬P, ¬Q
P_point, = ax.plot([], [], 'bo', markersize=15, label='P')
Q_point, = ax.plot([], [], 'go', markersize=15, label='Q')
notP_point, = ax.plot([], [], 'ro', markersize=15, label='¬P')
notQ_point, = ax.plot([], [], 'yo', markersize=15, label='¬Q')

# Flechas para las implicaciones
arrow_directa = ax.annotate('', xy=(0, 0), xytext=(0, 0), 
                           arrowprops=dict(arrowstyle='->', color='blue', lw=2))
arrow_contrar = ax.annotate('', xy=(0, 0), xytext=(0, 0), 
                           arrowprops=dict(arrowstyle='->', color='red', lw=2))

# Texto para las etiquetas
ax.text(2, 8, 'P', fontsize=12, ha='center', va='center')
ax.text(8, 8, 'Q', fontsize=12, ha='center', va='center')
ax.text(2, 2, '¬P', fontsize=12, ha='center', va='center')
ax.text(8, 2, '¬Q', fontsize=12, ha='center', va='center')

def init():
    # Posiciones iniciales de los puntos
    P_point.set_data([2], [8])
    Q_point.set_data([8], [8])
    notP_point.set_data([2], [2])
    notQ_point.set_data([8], [2])
    return P_point, Q_point, notP_point, notQ_point

def animate(i):
    # Dibujar flechas en diferentes pasos de la animación
    if i == 10:
        arrow_directa.set_position((2.2, 8))
        arrow_directa.xy = (7.8, 8)
    elif i == 30:
        arrow_contrar.set_position((7.8, 2))
        arrow_contrar.xy = (2.2, 2)
    return P_point, Q_point, notP_point, notQ_point, arrow_directa, arrow_contrar

# Crear y mostrar la animación
anim = FuncAnimation(fig, animate, init_func=init, frames=50, interval=200, blit=True)
plt.legend(loc='upper right')
plt.close()  # Evita que se muestre el gráfico estático antes de la animación
HTML(anim.to_jshtml())
````
